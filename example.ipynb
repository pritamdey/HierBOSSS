# HierBOSSS Example

In this notebook, we show the expression catching capability of HierBOSSS using the equation II_13_12 from the Feynmann Symbolic Regression Database (FSReD) across $5$ repetitions, $5$ parallel MCMC chains, and different noise levels $\sigma = \{0.00, 0.25, 1.00, 2.00\}$.

This equation relates to the **force on a moving charge in an electromagnetic field** and has the following symbolic form.

#### I_12_11: $F = q(E_f + vB\sin\theta)$

```python
from pyHierBOSSS_parallel import *
```

### Noiseless, $5$ repetitions, $10$ parallel MCMC chains

```python
I_12_11 = pd.read_csv("feynman_data/feynman_I_12_11.tsv", sep = "\t")
n = 3000
noise_sd = 0.00
K = 3
maxdepth = 4
move_weights = {
    "grow": 1.0,
    "prune": 1.0,
    "change_feature": 1.0,
    "change_operator": 1.0,
    "subtree_replace": 1.0,
    "delete_node": 1.0,
    "insert_node": 1.0
}

n_reps = 5
n_seeds = 10
top_k = 10

all_rep_results = []

for rep_idx in range(n_reps):
    print("\n" + "=" * 70)
    print(f"Repetition {rep_idx + 1} / {n_reps}")
    print("=" * 70)
    
    # repetition-specific data draw
    sample_seed = 10 + rep_idx
    noise_seed = 123 + rep_idx
    
    subset = I_12_11.sample(n=n, random_state=sample_seed)
    X_rep = subset.iloc[:, :-1].to_numpy(dtype=float)
    y_true_rep = subset.iloc[:, -1].to_numpy(dtype=float)
    
    np.random.seed(noise_seed)
    y_rep = y_true_rep + noise_sd * np.random.normal(size=n)
    
    ftset_rep = list(subset.columns[:-1])
    p_rep = X_rep.shape[1]
    
    # repetition-specific prior params
    alpha_op_rep = np.ones(len(opset))
    alpha_ft_rep = np.ones(len(ftset_rep))
    beta0_rep = np.zeros(K+1)
    V0_rep = 10.0 * np.eye(K+1)
    alpha = 0.95
    delta = 2.00
    a0 = 0.05
    b0 = 0.05
    
    prior_params_rep = (
        alpha_op_rep,
        alpha_ft_rep,
        alpha,
        delta,
        beta0_rep,
        V0_rep,
        a0,
        b0
    )
    
    wts_init_rep = [
        np.ones(len(opset)), np.ones(len(ftset_rep))
    ]
    wts_prop_rep = [
        np.ones(len(opset)), np.ones(len(ftset_rep))
    ]
    
    # instantiate MCMC object
    mcmc_rep = HierBOSSS_MCMC(
        X = X_rep, y = y_rep, K = K, maxdepth = maxdepth, prior_params = prior_params_rep,
        add_intercept = True, wts_init = wts_init_rep, wts_prop = wts_prop_rep, opset = opset,
        ftset = ftset_rep
    )
    
    # repetition-specific chain seeds
    seeds_rep = [123 * (rep_idx + 1) + i for i in range(n_seeds)]
    
    # run parallel MCMC chains
    mcmc_rep.run_parallel_chains(
        seeds=seeds_rep,
        move_weights=move_weights,
        maxiter=10_000,
        burnin=0,
        thin=1,
        n_jobs=n_seeds,
        show_progress=True
    )
    
     # ---------------------------------
    # summarize top-k across all chains
    # ---------------------------------
    expr_df_rep, beta_df_rep = mcmc_rep.summarize_parallel_topk_forests(top_k=top_k)

    print(f"\nTop {top_k} forests across all {n_seeds} chains for repetition {rep_idx + 1}:")
    display(expr_df_rep)
    print(expr_df_rep["tree_1"][0])
    print(expr_df_rep["tree_2"][0])
    print(expr_df_rep["tree_3"][0])

    print("Corresponding beta posterior means:")
    display(beta_df_rep)

    # optional trace plot for this repetition
    mcmc_rep.plot_parallel_log_jmp(alpha=0.5, linewidth=1.0)

    # ---------------------------------
    # store results
    # ---------------------------------
    all_rep_results.append({
        "rep_idx": rep_idx,
        "sample_seed": sample_seed,
        "noise_seed": noise_seed,
        "chain_seeds": seeds_rep,
        "X": X_rep,
        "y": y_rep,
        "y_true": y_true_rep,
        "ftset": ftset_rep,
        "mcmc": mcmc_rep,
        "expr_df": expr_df_rep,
        "beta_df": beta_df_rep
    })
```
